```python
import pandas as pd
from datetime import datetime
```

```python
d = datetime.today().strftime("%Y_%m_%d")
df = pd.read_csv(rf"..\data\interim\job_offers_{d}.csv", sep=";", encoding="UTF-8")
```

```python
df = df.query("salary_curr == 'PLN'")
```

```python
df['name']=df['name'].apply(lambda s: s.lower())
```

```python
df['location_city']=df['location_city'].apply(lambda s: s.lower())
```

```python
df['location_city'].unique()
```

```python
df.loc[df['location_city']=="warsaw", 'location_city'] = 'warszawa'
```

```python
df.loc[df['location_city']=="cracow", 'location_city'] = 'kraków'
```

```python
df['location_city'].unique()
```

```python
df['salary_avg'] = (df['salary_low'] + df['salary_high'])/2
```

```python
df['is_senior'] = df.apply(lambda o: 1 if 'senior' in o['name'] else 0, axis=1)
```

```python
df.loc[df['location_city']=="zdalna", 'location_city'] = 'N/A'
```

```python
display(df)
```

```python
df.to_csv(rf"..\data\processed\job_offers_{d}.csv", sep=";", encoding="UTF-8", index=False)
```
